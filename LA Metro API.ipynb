## LA Metro API

Do all of the imports and initialize variables

```python
from la_metro import LAMetro
import pandas as pd
import datetime
import timeit
import time
import numpy as np

client = LAMetro()

template_df = pd.DataFrame(columns = ['grouped time', 'actual time', 'last report', 'id', 'route', 'latitude', 'longitude', 'run'])

df = template_df #restore df

count = 0

df
```

I will begin to grab the data

Here are things to note:
    - A timer is created at the beginning to time the function
    
    - I create a temp_df which will hold all of the bus data
    
    - I set a secondary timer to run the while loop for whatever duration I choose
    
    - The while loop begins
    
    - I will try to get all of the bus data from the API, but if it doesn't I except with
      an empty list
      
    - Then I get the current time (which will make it easier to visualize),
      so I won't have to track every bus and update them accordingly
    
    - I grab the data with a try and except for most variables
    
    - Create a tuple from the data and append it to the temp_df
    
    - Once the time is up, then I append the temp_df to the regular df
    
    - I stop the timer and print the duration
    
    - I show what is being held in the df
    

```python
temp_df = template_df

#Your statements here

t_end = time.time() + (60 * 60 * 2) # <--- seconds * minutes * hours

print ("This will end at: ", time.asctime( time.localtime(t_end) ))

count = 0

while time.time() < t_end:
    
    print(len(temp_df))
    
    count = count + 1
    print('\n', "count: ", count)
    
    try:
        all_buses = client.bus.vehicles.all()
        print('\n', "num buses: ", len(all_buses))
    except:
        all_buses = []
        print('\n', "error")
    
    group_time = datetime.datetime.now()
    for bus in all_buses: #if error, then this loop will skip
        
        try:
            last_report = bus.seconds_since_report
        except:
            last_report = np.nan
        
        try:
            bus_id = bus.id
        except:
            bus_id = np.nan
        
        try:
            route = bus.route
        except:
            route = np.nan
        
        try:
            latitude = bus.latitude
        except:
            latitude = np.nan
        
        try:
            longitude = bus.longitude
        except:
            longitude = np.nan
        
        try:
            run = bus.run.name
        except:
            run = "none"
        
        new_bus = {'grouped time': group_time,
                   'actual time' : datetime.datetime.now(),
                   'last report' : last_report,
                   'id' : bus_id,
                   'route' : route,
                   'latitude' : latitude,
                   'longitude' : longitude,
                   'run' : run}
                   
                   
        #append row to the dataframe
        temp_df = temp_df.append(new_bus, ignore_index=True)
        
df = df.append(temp_df, ignore_index = True)

print("num iterations: ", count)

df
```
