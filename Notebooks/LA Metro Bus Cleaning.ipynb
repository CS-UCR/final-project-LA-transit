# LA Metro Bus Data Cleaning
I have data from the following days that need to be cleaned: 7/16, 7/17, 7/18, 7/20, 7/22

```python
import pandas as pd
```

```python
bus_df = pd.read_csv("../Bus Data/LA Metro Bus Data.csv")
bus_df
```

I need to change the grouped time and actual time to datetime format

```python
bus_df["grouped time"] = bus_df["grouped time"].apply(pd.to_datetime)
bus_df["actual time"] = bus_df["actual time"].apply(pd.to_datetime)
```

```python
bus_df.dtypes
```

I am also going to make an hour column

```python
bus_df["hour"] = bus_df["actual time"].map(lambda x: x.hour)
bus_df
```

I will now split the direction from the run column and make it its own column

```python
direction_df = bus_df["run"].str.split(' ', n = 1, expand = True)
direction_df.columns = ["direction", "name"]
bus_df["direction"] = direction_df["direction"]
bus_df
```

Here I am going to split the day from the actual time and make it its own column

```python
day_df = bus_df["actual time"].astype(str).str.split(' ', n = 1, expand = True)
day_df.columns = ["day", "time"]
bus_df["day"] = day_df["day"]
bus_df
```

I am going to change all of the day values into their day of the week values

```python
bus_df["day"].unique()
```

```python
bus_df['day'] = bus_df['day'].map({'2020-07-16' : 'Thursday',
                                   '2020-07-17' : 'Friday',
                                   '2020-07-18' : 'Saturday',
                                   '2020-07-20' : 'Monday',
                                   '2020-07-22' : 'Wednesday'
                                  })
bus_df
```

```python
bus_df.to_csv('../Bus Data/[Cleaned] LA Metro Bus Data.csv', header=True, index=False)
```

Now the data is cleaned and has been exported to csv, so let's move to the EDA portion
